#### Data Content
<blockquote>
<b>Company Name:</b> Name of Organization responsible for launch <br>
<b>Location:</b> String format of the launch location <br>
<b>Detail:</b> Specifics of each rocket <br>
<b>Status Rocket:</b> Showing if a rocket is currently in use <br>
<b>Rocket:</b> Cost of rocket used (in Millions USD) <br>
<b>Status Mission:</b> One of 4 categorical elements showing the the result of the launch<br>
<b>Country of Launch:</b> The country where the launch took place<br>
<b>Compagny's country of Origin:</b> The country that the organization is from <br>
<b>Compagny type:</b> The organization's category. Eg: SpaceX for private (P), NASA for Gov. institute/State (S)<br>
<b>DateTime:</b> Date and Time of the launch <br>
</blockquote>
