#### Data Content
<blockquote>
<b>Company Name:</b> Name of Organization responsible for launch <br>
<b>Location:</b> String format of the launch location <br>
<b>Detail:</b> Specifics of each rocket <br>
<b>Status Rocket:</b> Showing if a rocket is currently in use <br>
<b>Rocket:</b> Cost of rocket used (in Millions USD) <br>
<b>Status Mission:</b> One of 4 categorical elements showing the the result of the launch<br>
<b>Country of Launch:</b> The country where the launch took place<br>
<b>Compagny's country of Origin:</b> The country that the organization is from <br>
<b>Compagny type:</b> The organization's category. Eg: SpaceX for private (P), NASA for Gov. institute/State (S)<br>
<b>DateTime:</b> Date and Time of the launch <br>

## Python libraries

```python
# Importing Python libraries

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

## Read CSV files and Overview of the contents

```python
# Read CSV file
space_missions = pd.read_csv("Space_Corrected.csv")
print('Raw data :')
display(space_missions.head())

# Drop the two first columns that have been created during the loading of the file
space_missions = space_missions.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1)
print('After removing Unnamed: 0 and Unnamed: 0.1 columns :')
space_missions.head()
```

```python
#display the dataFrame space_missions
display(space_missions.head())
#print(space_missions.head())
```

## Data Visualization

```python
# Number of launches per company
fig, ax = plt.subplots()
ax = sns.countplot(data=space_missions,
                   x='Company Name',
                   order = space_missions['Company Name'].value_counts().index,
                  palette=sns.color_palette('Blues_r',56),
                  
                  )
ax.set_title('Number of launches per company')
ax.set_yscale('log')
plt.xticks(rotation=90)
plt.show()
```
